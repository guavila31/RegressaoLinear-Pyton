# Exercicio 1-A

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import os
```

```python
mes =    [1,2,3,4,5]
numero = [2508,2430,2791,2436,2652]


#aponta a fonte de dados "X" e "Y" Respectivamente - indica o icone que vai aparecer na tabela
plt.plot(mes,numero,'o')

#Titulo superior e legenda no eixo x e y
plt.title('Exercicio 1 - A')
plt.ylabel('Numero')
plt.xlabel('Mes')
plt.show()

dados = pd.DataFrame(zip(mes,numero), 
columns=['mes','numero'])
                   
dados
```

```python
X = dados['mes']
y = dados['numero']

X = sm.add_constant(X)
model= sm.OLS(y, X).fit()

model.summary()
```
